# Reading data
- data shape is num_of_images x image_size
- flattening the data images

```python
import cv2
import os
import glob
import numpy as np
import matplotlib.pyplot as plt

img_dir = "face" # Enter Directory of all images 
data_path = os.path.join(img_dir,'*pgm')
files = glob.glob(data_path)
data = np.zeros((361,2429))
index = 0
for f1 in files:
    img = cv2.imread(f1,cv2.IMREAD_GRAYSCALE)
    x = np.asarray(img)
    x = x.flatten()
    data[:,index] = x
    index=index+1
data = data.T
print(data.shape)
print(data)
```

# Display 10 random images

```python

num_columns = data.shape[0]
random_indices = np.random.choice(num_columns, size = 10, replace = False)
random_columns = data[random_indices,:]
print(random_columns.shape)

##Plot  images 
def disp_images(data, r, c):
    fig=plt.figure(figsize=(10, 10))
    columns = c
    rows = r
    for i in range(1, columns*rows +1):
        img = data[i-1,:].reshape(19,19)
        fig.add_subplot(rows, columns, i)
        plt.imshow(img, cmap='gray')
        plt.xticks([])
        plt.yticks([])
#display 10 images       
disp_images(random_columns, 5, 2)

```

# Building plain vanilla Autoencoder with encoding size 100

```python
import keras
from keras import layers
from sklearn.model_selection import train_test_split

class encoder(keras.Model):
    def __init__(self):
        super(encoder, self).__init__()
        self.dense1 = keras.layers.Dense(300,input_shape=(361,), activation='relu')
        self.dense2 = keras.layers.Dense(200,  activation='relu')
        self.dense3 = keras.layers.Dense(100,  activation='relu')

    def call(self, inputs):
        x = self.dense1(inputs)
        x = self.dense2(x)
        x = self.dense3(x)
        return x
    
class decoder(keras.Model):
    def __init__(self):
        super(decoder, self).__init__()
        self.dense1 = keras.layers.Dense(200,input_shape=(100,),  activation='relu')
        self.dense2 = keras.layers.Dense(300, activation='relu')
        self.dense3 = keras.layers.Dense(361, activation='relu')

    def call(self, inputs):
        x = self.dense1(inputs)
        x = self.dense2(x)
        x = self.dense3(x)
        return x
class AE(keras.Model):
    def __init__(self, encoder, decoder):
        super(AE, self).__init__()
        self.encoder = encoder
        self.decoder = decoder
    def call(self, inputs):
        encoded = self.encoder(inputs)
        decoded = self.decoder(encoded)
        return decoded

encoder = encoder()
decoder = decoder()

model = AE(encoder, decoder)
model.compile(optimizer='adam', loss='MSE')
```

# Normalizing the data by dividing gray scale images

```python
data = data/255
```

### epochs and batch size chosen after many trials

```python

train_x, test_x = train_test_split(data,test_size=0.3, random_state=42)
valid_x, test_x = train_test_split(test_x, test_size=0.5, random_state=42)

print(train_x.shape, valid_x.shape, test_x.shape)

model.fit(train_x, train_x,
                epochs=3000,
                batch_size=64,
                shuffle=True,
                validation_data=(valid_x, valid_x))
```
